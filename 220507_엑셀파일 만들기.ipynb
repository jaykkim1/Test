# 엑셀 연습

```python
import openpyxl
import random

# 기존 엑셀파일 열기
#wb = openpyxl.load_workbook('파일명.xlsx')

# 새로운 엑셀파일 생성
wb = openpyxl.Workbook()

# 시트 활성화
ws = wb.active
#ws = wb.get_sheet_by_name("Sheet1") # 엑셀 파일을 새로만들면 보통 첫번째 시트가 Active Sheet가 되므로 Worksheet 객체의 active 를 통해 현재 워크시트를 가져올 수 있지만, 
# 기 저장된 엑셀을 열면 이전 저장시 마지막에 선택된 시트를 활성화 하므로  wb.get_sheet_by_name('시트명') 을 사용하는 것이 안전하다(근데 파이참에서 오류)
# 혹은 index를 sheet이름으로하여 찾기(이게 더 좋다) --> wb['시트명'] 

# 시트이름 변경
ws.title = 'data' #이렇게 되면 엑셀파일에서 해당 시트명이 새롭게 지정된 이름으로 바뀐다. 

### 내용 채우기
# 헤더 추가
ws.append(['순번', '제품명', '가격', '수량', '합계']) # 엑셀의 append는 넣을때마다 리스트값들이 행단위로 추가된다. 딕셔너리에서의 리스트는 세로방향

# 데이터 추가 (랜덤으로 값 만들기)
name_list = ['기계식 키보드', '게이밍 마우스', '32인치 모니터', '마우스 패드']

for i in range(random.randint(5,10)): # 랜덤 정수(최소, 최대) --> 매출 데이터 (행): 5-10개
  name = random.choice(name_list)
  if name == "기계식 키보드":
    price = 12000
  elif name == "게이밍 마우스": #elif : 그렇지 않고 --라면
    price = 40000
  elif name == "32인치 모니터":
    price = 35000
  elif name == "마우스 패드":
    price = 20000
  ws.append([i+1, name, price, random.randint(1,5), f'=C{i+2}*D{i+2}']) #여기서 f스트링을 쓴것은 실제 해당 셀에 엑셀함수를 넣기 위함. i+2가 들어간것은 엑셀은 행이 1부터 시작인데다, 1행은 인덱스가 들어가서.



```

```python
import openpyxl
import random

# 새로운 엑셀파일 생성
total_wb = openpyxl.Workbook()

# 시트 활성화
total_ws = total_wb.active

# 시트이름 변경
total_ws.title = 'data' #이렇게 되면 엑셀파일에서 해당 시트명이 새롭게 지정된 이름으로 바뀐다. 

# 헤더 추가
total_ws.append(['순번', '제품명', '가격', '수량', '합계']) # 엑셀의 append는 리스트값들이 가로방향으로 나간다. 딕셔너리에서의 리스트는 세로방향

# 데이터 파일 로딩
file_list = ['쿠팡','11번가','스마트스토어']

for file in file_list:
  wb = openpyxl.load_workbook(f'03.엑셀자동화_{file}.xlsx', data_only=False) #data_only=True (값만, false -->값 및 수식)
  ws = wb.active
  for row in ws.iter_rows(min_row=2): #iter_rows()나 iter_cols()는 특정범위의 셀을 접근하는 방식임 (행이나 열을 탐색하는데 지정된 범위로,. 빈괄호이면 전체). 이자체는 generator에 불과. 이렇게 접근한 항목을 row에 나눠담음
  # [중요] 요약하자면 자료를 더미로 긁어서 (iter_rows()나 iter_cols()) --> 각 행이나 열을 뽑아다가(for row in ws.iter_rows()) --> 각 열 또는 행을 구성하는 각 셀을 뽑아내는(for cell in row) 작업
  # 여기서는 첫줄에 수동으로 설정한 인덱스가 들어가므로 최소값을 2로 잡는다. (엑셀의 줄 값(0이 아니라 1부터 시작, 0은 엑셀 인덱스 자리)을 준용)
    data =[]
    #print(row) # row는 튜플형태로서 각 행의 값(여러개의 셀데이터)들이 튜플형태로 포함되어 있어 (<Cell 'data'.A2>, <Cell 'data'.B2>, <Cell 'data'.C2>,,...) .value 형태로 뽑을수 없음. 그래서 이 값들을 개별적으로 뽑아서 분배시켜야 함. 
    for cell in row: #각 행의 셀값을 순차적으로 읽어오기. 
      data.append(cell.value) #넘에온 셀 데이터에 대해 값(.value)을 읽어서 data 빈 통에 붙여준다. 여기 for문 및 append를 통해 열또는 행 한줄이 완성. 상위 for문을 통해 모든 열 또는 행이 반복
    total_ws.append(data)



## 뒤죽박죽인 순번 새로 매기기 (방법 1)
# for row in total_ws.iter_rows(min_row=2, max_col=1): #스프레드시트의 1열(순번)을 새로 지정하는것이므로 반복탐색(iter_rows)기능을 쓰되 컬럼인덱스를 제외하므로 2행부터 끝까지, 열은 첫번째 열(순번)만 적용
#   for cell in row:
#     cell.value = row[0].row - 1 #[중요] 위에서 1열의 2행부터 끝까지 셀값을 읽어오기로 했는데, 여기서 row는 튜플형태이므로 인덱스를 적용할 수 있음. 값이 하나이므로 [0]을 붙임. 그리고 이때의 행인덱스 (.row)를 끌어다 넣게 지정
#  row는 (<Cell 'data'.A1>, <Cell 'data'.A2>,..) 이런식으로 나가는데 for문을 써서 하나씩 꺼내 cell에 넣는 셈임. 그리고 row[0]은 행마다 열이 하나이므로 그 개별 항목임. 그것의 열 인덱스를 반환(.row)하는것 

## 뒤죽박죽인 순번 새로 매기기 (방법 2)
i = 0
#print(total_ws['A'])
for cell in total_ws['A']: #A열 값들을 cell로 옮긴다. total_ws['A']자체는 암호화된 튜플형임 ((<Cell 'data'.A1>, <Cell 'data'.A2>,..). 열을 뽑아다가 셀로 넣어주는 작업
  if i !=0:
    cell.value = i #개별 데이터로 있기때문에 .value로 값을 지정할 수 있음
  i = i+1


# 엑셀로 저장
total_wb.save('03.엑셀자동화_total.xlsx')
```

```python
##### 전통적인 방식으로 만들어서 엑셀로 추출 #####
A=[]

# 헤더 추가
A.append(['순번', '제품명', '가격', '수량', '합계'])

# 데이터 추가
name_list = ['기계식 키보드', '게이밍 마우스', '32인치 모니터', '마우스 패드']

for i in range(random.randint(5,10)): # 랜덤 정수(최소, 최대) --> 매출 데이터 (행): 5-10개
  name = random.choice(name_list)
  if name == "기계식 키보드":
    price = 12000
  elif name == "게이밍 마우스":
    price = 40000
  elif name == "32인치 모니터":
    price = 35000
  elif name == "마우스 패드":
    price = 20000
  num=random.randint(1,5)
  A.append([i+1, name, price,num, price*num])

print(A)

import pandas as pd
print (pd.DataFrame(A))


# 데이터프레임으로 전환
df=pd.DataFrame(A)

# 엑셀로 저장
df.to_excel('test.xlsx')
```
